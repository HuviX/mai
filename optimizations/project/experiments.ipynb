```python
from optimizer import batchGD, Nadam
from utils import gen_dataset, plot_
from matplotlib import pyplot as plt
import numpy as np
```

```python
N_ITERS = 100000
ETA = 2e-3
BETA1  = 0.975
BETA2 = 0.999
EARLY_STOP = 1e-5 #Оптимизация завершается, если mse меньше этого значения
LR_DECAY_RATE = 0.95
bgd = batchGD(eta=ETA, decay_rate=LR_DECAY_RATE, early_stopping=1e-5)
nadam = Nadam(ETA, BETA1, BETA2, EARLY_STOP)
```

# Data_size = 30

```python
#Сгенерированный набор данных
data_size = 30
x_train, x_test, y_train, y_test, w = gen_dataset(data_size)


plt.plot(x_train, y_train)
plt.title("$f_w(x) = %d + %d*x + %d*x^2$" %(w[0], w[1], w[2]))
plt.grid();
```

## BatchGD (BatchSize = 1)

```python
%time
w_hat, hist, lr_hist, e = bgd.do_it(x_train, y_train, N_ITERS)
print("Полученные параметры", w_hat)

plot_(x_train, x_test, y_train, y_test, w_hat, hist)
plt.plot(np.arange(len(lr_hist)), lr_hist, label='lr-history')
plt.title("lr history")
plt.legend()
plt.grid()
```

## Nadam

```python
%time
w_hat, hist = nadam.do_it(x_train, y_train, N_ITERS)
print("Полученные параметры", w_hat)
plot_(x_train, x_test, y_train, y_test, w_hat, hist)
```

# Parameters comparison

```python
#Сгенерированный набор данных
data_size = 30
x_train, x_test, y_train, y_test, w = gen_dataset(data_size)


plt.plot(x_train, y_train)
plt.title("$f_w(x) = %d + %d*x + %d*x^2$" %(w[0], w[1], w[2]))
plt.grid();
```

```python
best_eta = 2e-3
best_beta1 = 0.975
best_beta2 = 0.999
etas = [0.001, 0.01, 0.05, 0.1, 1]
n_iters = 10000
eta_grid = {}
for i, eta_ in enumerate(etas):
    nadam = Nadam(eta=eta_, beta1=best_beta1, beta2=best_beta2, early_stopping=1e-7)
    _, hist = nadam.do_it(x_train, y_train, n_iters)
    key = (eta_, best_beta1, best_beta2, len(hist))
    eta_grid[key] = hist
    print(f"{i}/{len(etas)}")
```

```python
import matplotlib
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 19}
matplotlib.rc('font', **font)
for k in eta_grid:
    hist = eta_grid[k]
    plt.plot(list(range(len(hist))),hist, label="$ \\alpha = %s$" % (k[0]))
plt.grid()
plt.legend()
plt.rc('legend', fontsize=19)
plt.xlim([0, 1000])
plt.xlabel("n iterations")
plt.ylabel("loss")
```

```python
betas_1 = np.linspace(0.9, 0.999, 10)
beta1_grid = {}
for i, beta in enumerate(betas_1):
    nadam = Nadam(eta=best_eta, beta1=beta, beta2=best_beta2, early_stopping=1e-7)
    _, hist = nadam.do_it(x_train, y_train, n_iters)
    key = (best_eta, beta, best_beta2, len(hist))
    beta1_grid[key] = hist
    print(f"{i}/{len(betas_1)}")
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 19}
matplotlib.rc('font', **font)
for k in beta1_grid:
    hist = beta1_grid[k]
    plt.plot(list(range(len(hist))),hist, label="$ \\beta_1 = %s$" % (k[1]))
plt.grid()
plt.legend()
plt.rc('legend', fontsize=19)
plt.xlim([0, 1000])
plt.xlabel("n iterations")
plt.ylabel("loss")
```

```python
betas_2 = np.linspace(0.9, 0.999, 10)
beta2_grid = {}
for i, beta in enumerate(betas_2):
    nadam = Nadam(eta=best_eta, beta1=best_beta1, beta2=beta, early_stopping=1e-7)
    _, hist = nadam.do_it(x_train, y_train, n_iters)
    key = (best_eta, best_beta1, beta, len(hist))
    beta2_grid[key] = hist
    print(f"{i}/{len(betas_2)}")
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 19}

for k in beta2_grid:
    hist = beta2_grid[k]
    plt.plot(list(range(len(hist))),hist, label="$ \\beta_2 = %s$" % (k[2]))
plt.grid()
plt.legend()
plt.xlim([1000, 2000])
plt.ylim([0, 2])
plt.rc('legend', fontsize=19)
plt.xlabel("n iterations")
plt.ylabel("loss")
```
