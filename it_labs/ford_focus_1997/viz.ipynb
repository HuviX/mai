```python
import numpy as np
import matplotlib.pyplot as plt
import random
from utils import generate_points
from convex_hull import Convex_hull
```

```python
alg = Convex_hull('gift')
```

```python
size = 1000
mu = 0
sigma = 20
my_points = generate_points(size, mu, sigma)
```

```python
res_points = alg.find(my_points)
```

```python
fig, ax = plt.subplots(figsize=(20,20))
plt.scatter(my_points[:, 0], my_points[:, 1], c='b')
for i, _ in enumerate(res_points[:-1]):
    plt.plot([res_points[i][0], res_points[i+1][0]],[res_points[i][1], res_points[i+1][1]], c='r')
plt.grid()
```
