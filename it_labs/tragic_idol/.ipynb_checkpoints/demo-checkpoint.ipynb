```python
import pandas as pd
import numpy as np
from tree import tree
from IPython.display import Image


df = pd.read_csv("data.csv", sep=',')
df.reset_index(inplace=True, drop=True)
df
```

```python
target = 'text_editor'
path = 'my_tree.png'

#Создаем объект класса tree. Ну да, пустой конструктор и что?

my_tree = tree()
my_tree.fit(df, target)
my_tree.to_png(path)
```
